```python
import numpy as np
```

## Initialization

```python
def matrix_ini():
    m = int(input("Enter the number of rows:")) 
    n = int(input("Enter the number of columns:")) 
    
    print("Enter the entries of matrix in single line, separated by space: ")
    elements = list(map(int, input().split()))
    matrix = np.array(elements).reshape(m, n) 
    print(matrix,'\n') 
    
    return matrix
```

```python
m1 = matrix_ini()
m2 = matrix_ini()
```

## Multipilcation

```python
def m_mul(a,b):
    mul = np.zeros((len(a[0]),len(b)))
    for i in range(len(a)): 
        for j in range(len(b[0])): 
            for k in range(len(b)): 
                mul[i][j] += a[i][k] * b[k][j] 
    return mul
```

```python
if len(m1[0]) != len(m2):
    # Condition for multiplication
    print('Multiplication not possible.')
else:
    multiplication_m = m_mul(m1,m2)
    print(multiplication_m)
```

**Multiplication by numpy library function dot**

```python
# For checking of our created function
multiplication_n = np.dot(m1,m2)
print(multiplication_n)
```

```python
if multiplication_m.all() == multiplication_n.all():
    print("The multiplication is done correctly.")
    print("Multiplication of the matrices:\n", multiplication_n)
```

## Matrix solve

Let the area is divided as<br>


The length   along the  axis is divided into   equal segments, while the width   along the  axis is divided into   equal segments, hence giving<br>
$$\Delta x = \frac{L}{m}$$<br>
$$\Delta y = \frac{W}{n}$$<br>
L - length of grid<br>
W - width of grid<br>
Now we will apply the finite difference approximation of the partial derivatives at a general interior node ( ).<br>
$$\frac{\partial^{2}{T}}{\partial{x}^{2}}|_{i,j} \equiv \frac{T_{i+1,j} -2T_{i,j} + T_{i-1,j}}{(\Delta x)^{2}}$$<br>
$$\frac{\partial^{2}{T}}{\partial{y}^{2}}|_{i,j} \equiv \frac{T_{i,j+1} -2T_{i,j} + T_{i,j-1}}{(\Delta y)^{2}}$$

As both above equations are central divided difference approximations of the second derivatives. Substituting Equations  in the partial differential equation , we get<br>
$$\frac{T_{i+1,j} -2T_{i,j} + T_{i-1,j}}{(\Delta x)^{2}} + \frac{T_{i,j+1} -2T_{i,j} + T_{i,j-1}}{(\Delta y)^{2}} = 0$$<br>
For a grid with<br>
$$\Delta x = \Delta y$$<br>
the equation then can be simplified as<br>
$$T_{i+1,j} + T_{i-1,j} + T_{i,j+1} + + T_{i,j-1} - 4T_{i,j} = 0$$

Given,<br>
$T_{01} = T_{02} = 75$<br>
$T_{10} = T_{20} = 200$<br>
$T_{31} = T_{32} = 25$<br>
$T_{13} = T_{23} = 0$<br>

putting the values we get final linear equations as,<br>
$$-4T_{12} + T_{22} + T_{11} + 0.T_{21} = -225$$
$$T_{12} - 4T_{22} + 0.T_{11} + T_{21} = -25$$
$$T_{12} + 0.T_{22} - 4T_{11} + T_{21} = -275$$
$$0.T_{12} + T_{22} + T_{11} - 4T_{21} = -75$$

Giving us,<br>
$$\begin{bmatrix} -4&1&1&0 \\ 1&-4&0&1 \\ 1&0&-4&1 \\ 0&1&1&-4 \end{bmatrix}\begin{bmatrix} T_{12} \\ T_{22} \\ T_{11} \\ T_{21} \end{bmatrix} = \begin{bmatrix} -225 \\ -25 \\ -275 \\ -75\end{bmatrix}$$

```python
A = matrix_ini()
B = matrix_ini()
```

```python
# linalg.solve is the function of NumPy to solve a system of linear scalar equations 
sol = np.linalg.solve(A, B)
print("Solutions:\n",sol)
```
